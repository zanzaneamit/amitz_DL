Q.1 Write a Python Program to implement your own myreduce() function which works exactly like
Python's built-in function reduce()


```python
import operator
def myreduce(fn,l):
  for i in range(len(l)):
    if i == 0:
      a=l[i]
    else:
      b=l[i]
      a=fn(a,b)
  return a

def myreduce1(fn,l):
  itr = 0
  for i in l:
    if itr == 0:
      a=l[itr]
    else:
      b=l[itr]
      a=fn(a,b)
    
    itr = itr + 1

  return a

lis = [ 1, 3, 4, 10, 4 ]
print (myreduce1(lambda x,y:x+y,lis)) 

print (myreduce1(operator.add,["geeks","for","geeks"]))

t=(1,2,3,4,5,6)
print (myreduce(lambda x,y:x+y,t)) 
```

Q.4 Implement a function longestWord() that takes a list of words and returns the longest one.

```python
def longestWord(list1):
  l_word = ''
  l_word_len = 0
  for i in list1:
    if len(i) > l_word_len:
      l_word_len = len(i)
      l_word = i
  return l_word

# testing function
l1 = ['a', 'bb', 'bdbdbdbdbdbdbd','ccc', 'dddd']
print(longestWord(l1))
```

Q.5 Write a Python Program(with class concepts) to find the area of the triangle using the below
formula.

area = (s*(s-a)*(s-b)*(s-c)) ** 0.5

Function to take the length of the sides of triangle from user should be defined in the parent
class and function to calculate the area should be defined in subclass.

```python
class triangle:
  def __init__(self,a,b,c):
    self.a = a
    self.b = b
    self.c = c

class area_triangle(triangle):
  def __init__(self,a,b,c):
    super().__init__(a,b,c)

  def semiperimeter(self):
    self.s = (self.a + self.b + self.c)/2
    

  def area(self):
    self.semiperimeter()
    self.a = pow(self.s*(self.s - self.a)*(self.s - self.b)*(self.s - self.c),0.5)
    return self.a
     
# testing
t1 = area_triangle(2,3,4)
print(t1.area())
```

Q.6 Write a function filter_long_words() that takes a list of words and an integer n and returns the list
of words that are longer than n.

```python
def filter_long_words(list1, n):
  for i in list1:
    if len(i) > n:
      yield i

# testing function
l1 = ['a', 'bb', 'bdbdbdbdbdbdbd','ccc', 'dddd']
print([i for i in filter_long_words(l1,2)])
```

Q.7 Write a Python program using function concept that maps list of words into a list of integers
representing the lengths of the corresponding words.
    
    Hint: If a list [ ab,cde,erty] is passed on to the python function output should come as [2,3,4]
    Here 2,3 and 4 are the lengths of the words in the list.

```python
def map_chr_2_num(list1):
  l_len=[]
  for i in list1:
    l_len.append(len(i))
  return l_len

# testing function
l1 = ['a', 'bb', 'bdbdbdbdbdbdbd','ccc', 'dddd']
print(map_chr_2_num(l1))
```
